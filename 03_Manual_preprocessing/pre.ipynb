# Replace missing values
1. Categorical (Text) - Mode (most frequently occuring)
2. Continuous (Numerical) - Mean or Median

```python
df["AirBags"].value_counts()
```

```python
mode = df["AirBags"].mode()[0]
print(mode)
```

```python
df["Luggage.room"].mean()
```

```python
df["Luggage.room"].median()
```

```python
df.dtypes
```

```python
cat_cols = df.select_dtypes(include="object").columns.tolist()
cat_cols
```

```python
num_cols = df.select_dtypes(include="number").columns.tolist()
num_cols
```

```python
def replacer(df: pd.DataFrame):
    cat_cols = df.select_dtypes(include="object").columns.tolist()
    num_cols = df.select_dtypes(include="number").columns.tolist()
    for col in df.columns:
        if col in cat_cols:
            mode = df[col].mode()[0]
            df[col] = df[col].fillna(mode)
        elif col in num_cols:
            mean = df[col].mean()
            df[col] = df[col].fillna(mean)
    # Print missing values replaced
    print("Missing values replaced")
```

```python
df.isna().sum()
```

```python
replacer(df)
```

```python
df.isna().sum()
```

# High Cardinality columns - Categorical columns
Columns with most unique values are called high cardinality

```python
a = df.select_dtypes(include="object").nunique()
a
```

```python
len(df)
```

```python
cardinality = a / len(df)
cardinality
```

```python
high_card = cardinality[cardinality >= 0.9]
high_card
```

```python
high_card.index
```

```python
df = df.drop(columns = high_card.index)
df.head()
```

# Step 3 - Seperate X and Y(Weight)
id is a serial number
Weight ~ Remaing Features

```python
X = df.drop(columns = ["id", "Weight"])
X.head()
```

```python
Y = df["Weight"]
Y.head()
```

```python
X_cat = X.select_dtypes(include="object")
X_cat.head()
```

```python
X_num = X.select_dtypes(include="number")
X_num.head()
```

# Categorical - OneHotEncoder

```python
X_cat
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(
    handle_unknown="ignore", sparse_output=False, drop="first"
).set_output(transform="pandas")
ohe.fit(X_cat)
```

```python
X_cat_pre = ohe.transform(X_cat)
X_cat_pre.head()
```

# Numerical Cols - StandardScaler
mean = 0, std=1

```python
X_num
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().set_output(transform="pandas")
scaler.fit(X_num)
```

```python
X_num_pre = scaler.transform(X_num)
X_num_pre.head()
```

# Combine Categorical and Continuous features

```python
X_pre = X_num_pre.join(X_cat_pre)
X_pre.head()
```

# Train Test Split


# Train - Model Training 
# Test - Model Evaluation

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2, random_state=21)
```

# Preprocessing Complete

# Model Training - Train Data

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
coefs = pd.Series(model.coef_, index=xtrain.columns)
coefs
```

# Model Evaluation

```python
# R2 score train
model.score(xtrain, ytrain)
```

```python
# R2 score test
model.score(xtest, ytest)
```

```python
from sklearn.metrics import (
    root_mean_squared_error,
    mean_absolute_error,
    mean_absolute_percentage_error,
    r2_score
)

def evaluate_model(model, x, y):
    # Predict results for x
    ypred = model.predict(x)
    # Calculate metrics
    rmse = root_mean_squared_error(y, ypred)
    mae = mean_absolute_error(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred)
    r2 = r2_score(y, ypred)
    # Printing the results
    print(f"RMSE : {rmse:.2f}")
    print(f"MAE : {mae:.2f}")
    print(f"MAPE : {mape:.2%}")
    print(f"R2 : {r2:.2%}")
```

```python
print("Train Results : ")
evaluate_model(model, xtrain, ytrain)
```

```python
print("Test Result : ")
evaluate_model(model, xtest, ytest)
```
